# Model preparation and results

```python
car=pd.read_csv(r'D:\Project\Clean_data.csv')
```

```python
car
```

```python
car.columns
```

```python
car.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
car
```

```python
X = car.drop('Price',axis=1)
Y = car['Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
ohe = OneHotEncoder()


from sklearn.metrics import r2_score


ohe.fit(X[['name','company','fuel_type']])
column_tr = make_column_transformer((OneHotEncoder(categories = ohe.categories_) ,['name','company','fuel_type']),remainder = 'passthrough')

```

```python
new = []
for i in range(2000):
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_tr,lr)
    pipe.fit(X_train,Y_train)
    Y_pred = pipe.predict(X_test)
    new.append([r2_score(Y_test,Y_pred),i])
    print("Accuracy score :",r2_score(Y_test,Y_pred)," Random state :",i)
```

```python
max(new)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state = 1262)
lr = LinearRegression()
pipe = make_pipeline(column_tr,lr)
pipe.fit(X_train,Y_train)
Y_pred = pipe.predict(X_test)
n=pipe.predict(pd.DataFrame([['Suzuki Swift','Maruti',2009,100,'Petrol']],columns=['name','company','year','kms_driven','fuel_type']))
```
